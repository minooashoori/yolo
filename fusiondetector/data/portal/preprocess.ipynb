```python
def calculate_iou(box1, box2):
    x1_max = max(box1[0], box2[0])
    y1_max = max(box1[1], box2[1])
    x2_min = min(box1[2], box2[2])
    y2_min = min(box1[3], box2[3])

    intersection_area = max(0, x2_min - x1_max + 1) * max(0, y2_min - y1_max + 1)
    box1_area = (box1[2] - box1[0] + 1) * (box1[3] - box1[1] + 1)
    box2_area = (box2[2] - box2[0] + 1) * (box2[3] - box2[1] + 1)
    iou = intersection_area / float(box1_area + box2_area - intersection_area)
    return iou
```

```python
def nms(df, overlap_thresh=0.5):

    if df.empty:
        return df

    # Ensure the DataFrame has the required column
    if 'boxes' not in df.columns:
        raise ValueError("DataFrame must have a column 'boxes'")

    filtered_boxes_list = []  # Initialize a list to store 'filtered_boxes' for all rows

    for index, row in df.iterrows():
        boxes = row['boxes']
        filtered_boxes = []

        if len(boxes) == 0:
            filtered_boxes_list.append(filtered_boxes)
            continue
        if len(boxes) == 1:
            filtered_boxes_list.append(boxes)
            continue

        while len(boxes) > 0:
            current_box = boxes[0]
            
            ious = []
            for i in range(1, len(boxes)):
                iou = calculate_iou(current_box, boxes[i])
                ious.append(iou)

            # Remove repeated boxes within the same row
            filtered_boxes.append(current_box)
            boxes = [boxes[i] for i in range(1, len(boxes)) if ious[i - 1] < overlap_thresh]

        filtered_boxes_list.append(filtered_boxes)

    # Assign the list of 'filtered_boxes' to the DataFrame
    df['filtered_boxes'] = filtered_boxes_list

    return df
```

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
def plot_boxes(img_url, boxes):
    # download image with requests

    
    response = requests.get(img_url)
    img = Image.open(BytesIO(response.content))
    img = np.array(img)
    # Create figure and axes
    fig, ax = plt.subplots(1)
    # Display the image
    ax.imshow(img)
    # Create a Rectangle patch
    for box in boxes:
        rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=1, edgecolor='r', facecolor='none')
        # Add the patch to the Axes
        ax.add_patch(rect)
    # remove axis
    plt.axis('off')
    # make tight layout
    plt.tight_layout()
    plt.show()
    
```

```python
# plot 100 images in a grid
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

def plot_grid(df, box_col:str, n_rows=10, n_cols=10, offset=0, with_class=False):
    # Create figure and axes
    fig = plt.figure(figsize=(20, 20))
    gs1 = gridspec.GridSpec(n_rows, n_cols)
    gs1.update(wspace=0.025, hspace=0.05) # set the spacing between axes.
    for i in range(0, n_rows*n_cols):
        # i = i + 1 # grid spec indexes from 0
        ax1 = plt.subplot(gs1[i])
        ax1.set_xticklabels([])
        ax1.set_yticklabels([])
        ax1.set_aspect('equal')
        row = df.iloc[i+offset]
        img_url = row['image_url']
        boxes = row[box_col]
        response = requests.get(img_url)
        img = Image.open(BytesIO(response.content))
        img = np.array(img)
        ax1.imshow(img)
        for box in boxes:
            if with_class:
                box = box[1:]
            rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=1, edgecolor='r', facecolor='none')
            # Add the patch to the Axes
            ax1.add_patch(rect)
    plt.tight_layout()
    plt.show()
```

```python
import requests

def get_image_dimensions(url):
    try:
        # Send a GET request to the URL to retrieve a portion of the image
        response = requests.get(url, stream=True)
        
        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Read only the first 1024 bytes of the response content
            image_data = response.content[:1024]
            
            # Check if it's a valid image header (e.g., for JPEG)
            if image_data.startswith(b'\xff\xd8'):
                height = int.from_bytes(image_data[163:165], byteorder='big')
                width = int.from_bytes(image_data[165:167], byteorder='big')
                
                return width, height
            else:
                return None  # Not a valid image
        else:
            return None  # Failed to fetch metadata
        
    except Exception as e:
        return None  # Error occurred
```

```python
#  create a xyxyn column (normalized xyxy)
def normalize_boxes(boxes, width, height, with_class:bool):
    def normalize_box(box, width, height):
        return [round(box[0]/width, 4), round(box[1]/height, 4), round(box[2]/width, 4), round(box[3]/height, 4)]

    normalize_boxes = [normalize_box(box, width, height) for box in boxes] if not with_class else [ [box[0]] + normalize_box(box[1:], width, height) for box in boxes]
    return normalize_boxes
```

```python


def make_category_boxes(category, boxes):
    if isinstance(boxes, str):
        import ast
        boxes = ast.literal_eval(boxes)
    boxes_dict = []
    
    for box in boxes:
        boxes_dict.append([category, box[0], box[1], box[2], box[3]])
    return boxes_dict

def merge_boxes(logo_boxes, face_boxes):
    # combine the two lists
    if len(logo_boxes) == 0:
        return face_boxes
    if len(face_boxes) == 0:
        return logo_boxes
    
    boxes = logo_boxes + face_boxes
    return boxes



logos_df['face_xyxy'] = logos_df.apply(lambda x: make_category_boxes(0, x['face_xyxy']), axis=1)
logos_df['logo_xyxy'] = logos_df.apply(lambda x: make_category_boxes(1, x['xyxy']), axis=1)
logos_df['xyxy'] = logos_df.apply(lambda x: merge_boxes(x['logo_xyxy'], x['face_xyxy']), axis=1)
```

```python
# transform xyxy to yolo
logos_df['yolo'] = logos_df.apply(lambda x: transform_boxes(x['xyxy'], 'xyxy', 'yolo', True), axis=1)
logos_df['xywh'] = logos_df.apply(lambda x: transform_boxes(x['xyxy'], 'xyxy', 'xywh', True), axis=1)
```

```python
# normalise yolo
logos_df['yolon'] = logos_df.apply(lambda x: normalize_boxes(x['yolo'], x['width'], x['height'], True), axis=1)
```

```python
# make yolo annotations for training, create a column with the annotations in string format: class, x_center, y_center, width, height
def make_annotation(class_box):
    cls, x_c, y_c, w, h = class_box
    return f"{cls} {x_c} {y_c} {w} {h}\n"

def make_annotations(boxes):
    content = ""
    for box in boxes:
        content += make_annotation(box)
    # remove te last \n
    content = content[:-1]
    return content

logos_df['yolo_annotations'] = logos_df['yolon'].apply(lambda x: make_annotations(x))
logos_df['xywh_annotations'] = logos_df['xywh'].apply(lambda x: make_annotations(x))

# remove the quotation marks
logos_df['yolo_annotations'] = logos_df['yolo_annotations'].apply(lambda x: x.replace('"', ''))
logos_df['xywh_annotations'] = logos_df['xywh_annotations'].apply(lambda x: x.replace('"', ''))


# format these columns as string to avoid some wired behaviour when saving the csv file
logos_df['yolo_annotations'] = logos_df['yolo_annotations'].astype(str)
logos_df['xywh_annotations'] = logos_df['xywh_annotations'].astype(str)
```

```python
# filter out images with no annotations
logos_df = logos_df[logos_df['yolo_annotations'] != '']
```
